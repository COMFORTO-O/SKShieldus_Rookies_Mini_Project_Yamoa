```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns
import pandas as pd

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

batter_record_df = pd.read_csv('../../csv/batter_record_data.csv', encoding='utf-8').head(15)
bowler_record_df = pd.read_csv('../../csv/bowler_record_data.csv', encoding='utf-8').head(15)

# 타자 그래프
sns.barplot(y='선수명', x='타율', data=batter_record_df, hue='타율')
plt.title('타자 그래프')
plt.show()

# 투수 그래프
sns.barplot(y='선수명', x='평균자책점', data=bowler_record_df, hue='평균자책점')
plt.title('투수 그래프')
plt.show()
```
