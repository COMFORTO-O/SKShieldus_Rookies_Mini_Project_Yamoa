```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns
import pandas as pd

#  
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

# CSV  
batter_record_df = pd.read_csv('../../csv/scrap_batter_record_data.csv', encoding='utf-8').head(15)
bowler_record_df = pd.read_csv('../../csv/scrap_bowler_record_data.csv', encoding='utf-8').head(15)

#  
plt.figure(figsize=(10, 6))
ax = sns.barplot(y='', x='', data=batter_record_df, color='skyblue')
plt.title(' ')

#    
for i, v in enumerate(batter_record_df['']):
    plt.text(v + 0.005, i, f'{v:.3f}', va='center')  #  ,  3

plt.legend([], [], frameon=False)  #  
plt.tight_layout()
plt.show()

#  
plt.figure(figsize=(10, 6))
ax = sns.barplot(y='', x='', data=bowler_record_df, color='salmon')  # hue 
plt.title(' ')

#    
for i, v in enumerate(bowler_record_df['']):
    plt.text(v + 0.1, i, f'{v:.2f}', va='center')

plt.legend([], [], frameon=False)
plt.tight_layout()
plt.show()
```
